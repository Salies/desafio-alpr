```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
features = np.load('features.npy')
labels = np.load('targets.npy')
```

```python
label_freq = np.unique(labels, return_counts=True)
label_freq = dict(zip(label_freq[0], label_freq[1]))
label_freq
```

```python
plt.figure(figsize=(10, 5))
# plotando quantidade de amostras por classe
plt.bar(label_freq.keys(), label_freq.values())
plt.xlabel('Caracteres')
plt.ylabel('Quantidade de amostras')
plt.title('Distribuição de caracteres no dataset')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, stratify=labels, random_state=6661)
```

```python
train_freq = np.unique(y_train, return_counts=True)
train_freq = dict(zip(train_freq[0], train_freq[1]))
test_freq = np.unique(y_test, return_counts=True)
test_freq = dict(zip(test_freq[0], test_freq[1]))
```

```python
# plot train test frequency
plt.figure(figsize=(10, 5))
plt.bar(train_freq.keys(), train_freq.values(), label='Treino')
plt.bar(test_freq.keys(), test_freq.values(), label='Teste')
plt.xlabel('Caracteres')
plt.ylabel('Quantidade de amostras')
plt.title('Distribuição de caracteres no dataset (Treino vs Teste)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# salvando os dados
np.save('features_train.npy', X_train)
np.save('features_test.npy', X_test)
np.save('targets_train.npy', y_train)
np.save('targets_test.npy', y_test)
```
